```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#import airbnb data from the csv file
df = pd.read_csv('airbnb_dataset.csv')
```

```python
# converet variable to datetime format
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
```

```python
df.info()
```

```python
# how many null rows for 'host_since'
len(df.loc[df['host_since'].isna()])
```

```python
# added end_date variable so calculation can be done for how long an accomdations have been hosted on airbnb
df['end_date'] = pd.to_datetime('31-12-2020')
```

```python
# calcualted reviews_per_month for each listing to fill gaps in the data.
df['reviews_per_month'] = (df['end_date'].dt.year - df['host_since'].dt.year) * 12 + (df['end_date'].dt.month - df['host_since'].dt.month)
```

```python
df.info()
```

```python
df.loc[df['reviews_per_month'].isna()]
```

```python
# dropping rows with missing information after exploring.
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
# changing type to boolean
df['host_is_superhost'] = df['host_is_superhost'].replace('t', '1')
```

```python
df['host_is_superhost'] = df['host_is_superhost'].replace('f', '0')
```

```python
df['host_is_superhost'] = pd.to_numeric(df['host_is_superhost'])
```

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)}, figsize=(15,10))
#df2 = df.loc[df['price'] <= 1500]
df2 = df.loc[df['price'] <= 1500]
mean=df2['price'].mean()
median=df2['price'].median()
mode=df2['price'].mode().get_values()[0]

sns.boxplot(df2["price"], ax=ax_box)
ax_box.axvline(mean, color='b', linestyle='--')
ax_box.axvline(median, color='r', linestyle='-')
ax_box.axvline(mode, color='g', linestyle='-')

sns.distplot(df2["price"], ax=ax_hist)
ax_hist.axvline(median, color='r', linestyle='--')
ax_hist.axvline(mode, color='g', linestyle='-')
ax_hist.axvline(mean, color='b', linestyle='-')


plt.legend({'Mean':mean,'Median':median,'Mode':mode})

ax_box.set(xlabel='')
plt.show()
```

```python
#Plot the mean price distribution by State and Room Type
sns.set_theme(style="whitegrid")


plt.figure(figsize=(15,10))
# Draw a nested barplot of Price by State and Room Type
g = sns.catplot(
    data=df, kind="bar",
    x="State", y="price", hue="room_type",palette="dark", alpha=.6, height=9
)
g.despine(left=True)
g.set_axis_labels("", "Price")
g.legend.set_title("")
```
